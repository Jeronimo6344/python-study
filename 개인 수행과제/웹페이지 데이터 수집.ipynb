```python
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
with requests.Session() as session:
    session.headers.update({
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    })
    url='https://data.hossam.kr/py/myfood.html'
    r = session.get(url)

    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)
```

```python
r.encoding='utf-8'
print(type(r.text))
r.text
```

```python
soup=BeautifulSoup(r.text)
print(type(soup))
soup
```

```python
mylist=[]

# 메뉴명
menu_name=[]
myselect1=soup.select('.food-content>h2')
for i in myselect1:
    menu_name.append(i.text.strip())
# menu_name

# 메뉴 설명
menu_info=[]
myselect2=soup.select('.food-content>p')
for i in myselect2:
    menu_info.append(i.text.strip())
# menu_info

# 메뉴 이미지
menu_image=[]
myselect3=soup.select('.img-wrapper>img')
for i in myselect3:
    menu_image.append(i['src'])
# menu_image

for i in range(0,len(menu_name)):
    processes=[]
    processes.append(menu_name[i])
    processes.append(menu_info[i])
    processes.append(menu_image[i])
    mylist.append(processes)
```

```python
new_resultset=[]

for r in mylist:
    item_dict={
        '메뉴명':r[0],
        '메뉴 설명':r[1],
        '메뉴 이미지':r[2]
    }

    new_resultset.append(item_dict)

df=DataFrame(new_resultset)
df.to_excel('MyFood.xlsx')
df
```

```python
def download(url,target):
    session.headers.update({
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    })

    try:
        r=session.get(url,stream=True)
        r.encoding='utf-8'

        with open(target,'wb') as f:
            f.write(r.raw.read())
            print(target,'가 저장되었습니다.')
    except Exception as e:
        print(target,'저장실패',e)
```

```python
import os

dirname='MyFood_img'
os.mkdir(dirname)

for i, v in enumerate(mylist):
    file_path=os.path.join(dirname,'%d.jpg'%i)
    download(v[2],file_path)
```
