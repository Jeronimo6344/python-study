```python
import requests
import json
```

```python
url='https://api.vworld.kr/req/address'
API_KEY='F984B5C8-641F-37F2-A768-A53BEA49B579'
```

```python
pp=[]
with open("서울교통공사_역주소 및 전화번호_20250318.csv",'r',encoding='euc-kr') as f:
    csv_list=f.readlines()
    for i, line in enumerate(csv_list):
        if i==0:
            continue
        k=line.strip().split(',')
        pp.append(k[5])
```

```python
data=[]
with requests.Session() as session:
    session.headers.update({
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    })
    for j in range(0,len(pp)):
        r=session.get(url,params={'address':pp[j],'key':API_KEY, 'request':'getCoord','type':'road','service':'address'})
        r.encoding='utf-8'
        mydict=r.json()
        try:
            data.append(mydict['response']['result']['point'])
        except Exception as e:
            data.append({'x':'None','y':'None'})
            continue

        if r.status_code != 200:
            msg='[%d Error] %s 에러가 발생함'%(r.status_code,r.reason)
            raise Exception(msg)
```

```python
with open("서울교통공사_역주소 및 전화번호_20250318.csv",'r',encoding='euc-kr') as ff:
    csv_list1=ff.readlines()
    with open("subway_stations_with_coords.csv",'w',encoding='euc-kr') as fff:
        for h in range(0,len(pp)):
            fff.write(csv_list1[h].strip())
            if h==0:
                fff.write(',latitude,longitude')
            if h!=0:
                fff.write(',')
                fff.write(data[h-1]['y'])
                fff.write(',')
                fff.write(data[h-1]['x'])
            if h!=len(pp):
                fff.write('\n')
```
