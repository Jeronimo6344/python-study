```python
import requests
from pandas import DataFrame
```

```python
url='https://dapi.kakao.com/v3/search/book'

query='파이썬'
page=1
size=50

# 헤더에 'Authorization':'KakaoAK ___' 꼴로 입력
key='61ae79e2471fc01ae18f0160c7506ef1'
```

```python
resultset=[]

with requests.Session() as session:
    session.headers.update({
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
        'Authorization':'KakaoAK %s'%key
    })

    for i in range(1,51):           # 반복문 돌려서 모든 페이지 탐색
        r=session.get(url,params={'query':query,'page':i,'size':size})     # GET 방식에는 딕셔너리 형태로 params 제공

    # 정상코드 출력이 안되면 에러 메세지만 출력하고 다음으로 넘어가기
        if r.status_code != 200:
            msg='[%d Error] %s 에러가 발생함'%(r.status_code,r.reason)
            print(msg)
            continue

        print('%d page 데이터 수신 완료 >> %s'%(i,r))
        r.encoding='utf-8'
        mydict=r.json()
        resultset.extend(mydict['documents'])

print('총 %d건의 데이터가 수집됨'%len(resultset))
```

```python
df=DataFrame(resultset)
df.to_excel('카카오 책 검색(%s)-전체 데이터.xlsx'%query)
```

```python
def download(url,target):
    session.headers.update({
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    })

    try:
        r=session.get(url,stream=True)
        r.encoding='utf-8'

        with open(target,'wb') as f:
            f.write(r.raw.read())
            print(target,'가 저장되었습니다.')
    except Exception as e:
        print(target,'저장실패',e)
```

```python
import os
import datetime as dt

dirname=dt.datetime.now().strftime('%Y%m%d-%H%M%S')
os.mkdir(dirname)

for i, v in enumerate(resultset):
    file_path=os.path.join(dirname,'%d.jpg'%i)
    download(v['thumbnail'],file_path)
```

```python
from concurrent import futures

dirname=dt.datetime.now().strftime('%Y%m%d-%H%M%S')
os.mkdir(dirname)

with futures.ThreadPoolExecutor() as executor:
    for i, v in enumerate(resultset):
        file_path=os.path.join(dirname,'%d.jpg'%i)
        executor.submit(download,v['thumbnail'],file_path)
```
