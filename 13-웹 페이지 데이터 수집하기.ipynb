# HTML 데이터 추출(1)

```python
import requests
from bs4 import BeautifulSoup
```

```python
with requests.Session() as session:
    session.headers.update({
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    })
    url='https://data.hossam.kr/py/sample.html'
    r = session.get(url)

    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
    
    print(r)
```

```python
r.encoding='utf-8'
print(type(r.text))
r.text
```

```python
soup=BeautifulSoup(r.text)
print(type(soup))
soup
```

```python
myselect=soup.select('h1')
print(type(myselect))
myselect
```

```python
mytag=myselect[0]
print(type(mytag))
mytag
```

```python
mytext=mytag.text.strip()
mytext
```

```python
myselect=soup.select('.myclass')
myselect
```

```python
for i, v in enumerate(myselect):
    print('%d번째 요소 : %s'%(i,v.text.strip()))
```

```python
myli=myselect[2].select('li')
myli
```

```python
for i in myli:
    print(i.text.strip())
```

```python
myselect=soup.select('a[href]')
myselect
```

```python
for i,v in enumerate(myselect):
    print('----[%d]----'%i)
    print(v.attrs)

    if 'href' in v.attrs:
        print('%d번째의 href속성값 : %s'%(i,v.attrs['href']))
```
